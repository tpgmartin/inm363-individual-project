```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_style()
%matplotlib inline
```

```python
act_sims = pd.read_csv('../cosine_similarities/concept_acts/concept_cosine_similarities.csv')
```

```python
act_sims.head()
```

```python
act_sims.dtypes
```

```python
act_sims['label_1'] = act_sims['concept_1'].apply(lambda c: '_'.join(c.split('_')[:-1]))
act_sims['label_2'] = act_sims['concept_2'].apply(lambda c: '_'.join(c.split('_')[:-1]))
```

```python
pd.pivot_table(act_sims, values='cosine_similarity', index=['label_1'], columns=['label_2'], aggfunc=[np.mean, np.std])
```

```python
pd.pivot_table(act_sims, values='cosine_similarity', index=['label_1'], columns=['label_2'], aggfunc=[np.max, np.min])
```

```python
act_sims[(act_sims['label_1'] == 'ambulance') & (act_sims['label_2'] == 'ambulance')].groupby(['concept_1', 'concept_2'])['cosine_similarity'].agg([np.mean, np.std]).reset_index().sort_values(by=['mean','std'], ascending=False)
```

```python
act_sims[(act_sims['label_1'] == 'ambulance') & (act_sims['label_2'] == 'jeep')].groupby(['concept_1', 'concept_2'])['cosine_similarity'].agg([np.mean, np.std]).reset_index().sort_values(by=['mean','std'], ascending=False)
```

```python
data = act_sims[(act_sims['label_1'] == 'ambulance')][['label_2', 'cosine_similarity']].sort_values(by=['label_2', 'cosine_similarity'])
```

```python
plot_data = [(label, data[data['label_2'] == label]['cosine_similarity'].values.tolist()) for label in  data['label_2'].unique()]

labels = [data[0] for data in plot_data]
act_sim_data = [data[1] for data in plot_data]

plt.figure(figsize=(14,8))
plt.boxplot(act_sim_data, labels=labels)
plt.show()
```

```python
act_sims[(act_sims['label_1'] == 'ambulance') & (act_sims['label_2'] != 'ambulance')].groupby(['concept_1', 'concept_2'])['cosine_similarity'].agg([np.mean, np.std]).reset_index().sort_values(by=['mean','std'], ascending=False)
```
