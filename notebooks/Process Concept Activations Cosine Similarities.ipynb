```python
avg_concept_sims_by_label = []

for concept in act_sims['concept_1'].unique():
    avg_sims = act_sims[(act_sims['concept_1'] == concept) & (act_sims['concept_1'] != act_sims['concept_2'])].groupby(['concept_1','concept_2'])['cosine_similarity'].agg([np.mean, np.std, 'count'])
    avg_sims.reset_index(inplace=True, drop=False)
    avg_sims.sort_values(by=['concept_1', 'mean', 'std', 'count'], ascending=False, inplace=True)
    avg_concept_sims_by_label.append(avg_sims)
```

```python
tmp = pd.concat(avg_concept_sims_by_label)
tmp.to_csv('../cosine_similarities/average_acts_sims_by_concept.csv', index=False)
```

```python
len(avg_concept_sims_by_label)
```

```python
concept_1 = []
concept_2 = []
mean = []
std = []
concept_means = []
concept_stds = []

for df in avg_concept_sims_by_label:

    concept_mean = np.mean(df['mean'])
    concept_std = np.mean(df['std'])

    for idx in [0, -1]:

        concept_1.append(df.iloc[idx,0])
        concept_2.append(df.iloc[idx,1])
        mean.append(df.iloc[idx,2])
        std.append(df.iloc[idx,3])
        concept_means.append(concept_mean)
        concept_stds.append(concept_std)


min_max_only = pd.DataFrame({
    'concept_1': concept_1,
    'concept_2': concept_2,
    'mean': mean,
    'std': std,
    'concept_mean': concept_means,
    'concept_std': concept_stds
})

min_max_only.to_csv('../cosine_similarities/min_max_acts_sims_by_concept.csv', index=False)
```
