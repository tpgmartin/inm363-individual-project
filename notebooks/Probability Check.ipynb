```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
n = 1000
p = (100*100)/(400*600)
q = 1 - p
```

```python
mu = n*p
sigma = np.sqrt(n*p*q)
```

```python
no_times_selected = np.random.normal(mu, sigma, n)
```

```python
plt.hist(no_times_selected)
```

```python
no_times_selected_updated = []
for x in no_times_selected:
    if x < 0:
        no_times_selected_updated.append(0)
    else:
        no_times_selected_updated.append(np.round(x))
```

```python
pixel_values = [-1,0,1]
```

```python
values = []
for x in no_times_selected_updated:
    values.append(x * random.choice(pixel_values))
```

```python
plt.hist(values)
```

```python
np.mean(values)
```

```python
np.std(values)
```
