```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_style()
%matplotlib inline
```

```python
cav_sims = pd.read_csv('../cosine_similarities/concept_cavs/concept_cosine_similarities.csv')
```

```python
cav_sims.head()
```

```python
cav_sims.dtypes
```

```python
cav_sims['label_1'] = cav_sims['concept_1'].apply(lambda c: '_'.join(c.split('_')[:-1]))
cav_sims['label_2'] = cav_sims['concept_2'].apply(lambda c: '_'.join(c.split('_')[:-1]))
```

```python
pd.pivot_table(cav_sims, values='cosine_similarity', index=['label_1'], columns=['label_2'], aggfunc=[np.mean, np.std])
```

```python
pd.pivot_table(cav_sims, values='cosine_similarity', index=['label_1'], columns=['label_2'], aggfunc=[np.max, np.min])
```

```python
cav_sims[(cav_sims['label_1'] == 'ambulance') & (cav_sims['label_2'] == 'ambulance')].groupby(['concept_1', 'concept_2'])['cosine_similarity'].agg([np.mean, np.std]).reset_index().sort_values(by=['mean','std'], ascending=False)
```

```python
cav_sims[(cav_sims['label_1'] == 'ambulance') & (cav_sims['label_2'] == 'jeep')].groupby(['concept_1', 'concept_2'])['cosine_similarity'].agg([np.mean, np.std]).reset_index().sort_values(by=['mean','std'], ascending=False)
```

```python
data = cav_sims[(cav_sims['label_1'] == 'ambulance')][['label_2', 'cosine_similarity']].sort_values(by=['label_2', 'cosine_similarity'])
```

```python
plot_data = [(label, data[data['label_2'] == label]['cosine_similarity'].values.tolist()) for label in  data['label_2'].unique()]

labels = [data[0] for data in plot_data]
cav_sim_data = [data[1] for data in plot_data]

plt.figure(figsize=(14,8))
plt.boxplot(cav_sim_data, labels=labels)
plt.show()
```
