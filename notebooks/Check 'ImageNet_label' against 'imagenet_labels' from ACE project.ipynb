```python
ace_imagenet_labels = pd.read_csv('../labels/imagenet_labels.txt', index_col=False, names=['labels'])
ace_imagenet_labels = ace_imagenet_labels[1:]
ace_imagenet_labels.reset_index(drop=True, inplace=True)
ace_imagenet_labels['line_no'] = ace_imagenet_labels.index.values + 1

ace_imagenet_labels
```

```python
imagenet_labels = pd.read_csv('../labels/ImageNet_label.txt', index_col=False, sep='\s{4}', names=['img_train_folder', 'labels'], engine='python')
imagenet_labels['line_no'] = imagenet_labels.index.values + 1
imagenet_labels
```

```python
labels = ace_imagenet_labels.merge(imagenet_labels, on='line_no')
labels
```

```python
labels.loc[:, 'labels_y'] = labels['labels_y'].apply(lambda l: l.split(', '))
```

```python
for idx, row in labels.iterrows():
    
    if row['labels_x'] not in row['labels_y']:
        print(row)
```
