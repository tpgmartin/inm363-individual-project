This notebook is to map the ILSVRC2012 validation images to the labels found in `https://github.com/duanyzhi/ImageNet_Label`

```python
from glob import glob
import pandas as pd
```

```python
validation_class_label_line_nos = pd.read_csv('../labels/ILSVRC2012_validation_ground_truth.txt', sep=' ', names=['line_no'])
validation_class_label_line_nos['img_no'] = validation_class_label_line_nos.index.values + 1
validation_class_label_line_nos
```

```python
imagenet_labels = pd.read_csv('../labels/ImageNet_label.txt', index_col=False, sep='\s{4}', names=['img_train_folder', 'labels'], engine='python')
imagenet_labels['line_no'] = imagenet_labels.index.values + 1
imagenet_labels
```

```python
validation_class_labels = validation_class_label_line_nos.merge(imagenet_labels, on='line_no')
validation_class_labels.sort_values(by='img_no', inplace=True)
validation_class_labels.reset_index(drop=True, inplace=True)
validation_class_labels
```

```python
validation_class_labels['img_file_name'] = validation_class_labels['img_no'].apply(lambda x: f'ILSVRC2012_val_{x:08}.JPEG')
```

```python
validation_class_labels = validation_class_labels[['img_file_name', 'img_train_folder', 'labels']]
```

```python
validation_class_labels.to_csv('../labels/validation_class_labels.csv', index=False)
```
